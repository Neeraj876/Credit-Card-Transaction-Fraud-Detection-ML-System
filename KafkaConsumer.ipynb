```python
import pandas as pd
import os
from kafka import KafkaConsumer
from time import sleep
from json import loads
import json
import boto3
```

```python
consumer = KafkaConsumer('transformed_data', bootstrap_servers=['107.20.106.196:9092'])
```

```python
for c in consumer:
    print(c.value)
```

```python
# s3 = s3fs.S3FileSystem()
s3 = boto3.client('s3')
```

```python
response = s3.list_buckets()  # List all buckets to confirm connection
print(response)
```

```python
bucket_name = 'credit-card-transaction-processed-data'

for count, message in enumerate(consumer):
    print(f"Message {count}: {message.value}")
    
    # Construct the S3 object key (file name)
    s3_key = f"credit_card_transaction_data_{count}.json"
    
    # Upload the message to S3 as a JSON file
    s3.put_object(
        Bucket=bucket_name,
        Key=s3_key,
        Body=json.dumps(message.value),  # Convert the message value to JSON string
        ContentType='application/json'  # Optional: Set the content type
    )
    
    print(f"Uploaded to S3: {s3_key}")
```
